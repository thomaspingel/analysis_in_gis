```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
```

```python
url = 'https://raw.githubusercontent.com/thomaspingel/geodata/master/county_health_rankings/chr_2019.csv'
df = pd.read_csv(url,dtype={'FIPS':'str'})
```

```python
df.columns.values
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(np.sqrt(df['Demographics - Population']),df['Low birthweight - % LBW'],'.')
plt.xlabel('population (sqrt)')
plt.ylabel('% Low Birth Weight')

plt.subplot(2,2,2)
plt.plot(np.sqrt(df['Demographics - Population']),df['Flu vaccinations - % Vaccinated'],'.')
plt.xlabel('population (sqrt)')
plt.ylabel('Flu vaccinations - % Vaccinated')

plt.subplot(2,2,3)
plt.plot(np.sqrt(df['Demographics - Population']),df['Child mortality - Child Mortality Rate'],'.')
plt.xlabel('population (sqrt)')
plt.ylabel('Child mortality - Child Mortality Rate')

plt.subplot(2,2,4)
plt.plot(np.sqrt(df['Demographics - Population']),df['Demographics - % Not Proficient in English'],'.')
plt.xlabel('population (sqrt)')
plt.ylabel('% Not Proficient in English')
```
