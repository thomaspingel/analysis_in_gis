#### Key Steps:

1) Build a query to get the data we are interested in

2) Use urllib to encode the query string. 
  * One reason we do this is that you cannot pass spaces and other special characters to a url, by encoding it we have a string with url friendly parameters.

3) Use urllib to send the request to the server and return the response. 
  * Urllib has all the functionality to confirm if a request was successful or not and handle the passing of the data.
  
4) Receive the response from the server and parse the response string as a json to begin processing.

### Make a request to the API

```python
# Import the required libraries from urllib
from urllib import request
from urllib import parse
```

```python
# The url of the Analzye Boston datastore API
url = "https://data.boston.gov/api/3/action/datastore_search_sql"
```

We will now build the query for the API. If you are not familiar with SQL you can move on without worrying about the nitty gritty of the query.

The Crime Incident Reports can be accessed using the identifier `12cb3883-56f5-47de-afa5-3b1cf61b257b`. To select crimes of the assault type we are using the column `OFFENSE_CODE_GROUP` (you can see the columns from the [API URL](https://data.boston.gov/dataset/crime-incident-reports-august-2015-to-date-source-new-system)). We are going to query this column for the values `Aggravated Assault` or `Simple Assault`. Using a `SQL` `AND` statement we also add in a filter for the year `2017`. You can change the query values to analyze any crime types over any period that you are interested in. This is one of the great features of connecting to a live data API, we can change our analysis without having to have downloaded all of the data at once.

We create a dictionary with a key `sql`, this key is used by the API to accept SQL query paremeters. The query string is then encoded for sending in a http request.

```python
# Build SQL query string
params = {
    'sql': """SELECT * from "12cb3883-56f5-47de-afa5-3b1cf61b257b" WHERE 
    ("OFFENSE_CODE_GROUP" = 'Aggravated Assault' OR "OFFENSE_CODE_GROUP" = 'Simple Assault') 
    AND "YEAR"='2017'"""
}
query_string = parse.urlencode(params)

print("This is the query string, spaces and special characters have been replaced\n\n{}".format(query_string))
```

```python
# Concatinate the url and the query string with a question mark
full_url = url + "?" + query_string
print(full_url)
```

Now we send the request to the server. If we get a response code of **200** the request was a success. If it did not work try again as the server sometimes errors when responding. If it does not work after several tries confirm that the server is loaded and that the Crime Incident Reports datasource item id has not changed


```python
# Request data from server
try:
    resp = request.urlopen(full_url)
    if resp.code==200:
        print("Response Code {} -- it was a success".format(resp.code))
    else:
        print("Response Code {} -- something went wrong, try the request again".format(resp.code))
except Exception as e:
    print(str(e))
```

### Clean the data

We will now take a look at the data and do some data cleaning: change data format, add timestamp field, and remove data with no or invalid location information.

We want to use the coordinates from the Lat and Long columns to create points. If there are crime incidents without location coordinates we will drop them. We will assess how many records we lose due to a lack of coordinates. Depending on the analysis requirements, dropping crime incidents without coordinates may not be acceptable.

```python
# Most fields are objects in the dataframe, we will convert the following fields to numeric values:
# Lat and Long need to be numeric for creating SeDF
# HOUR, MONTH, YEAR are for analysis purpose
numeric_fields = ['Lat', 'Long', 'HOUR', 'MONTH', 'YEAR']
for field in numeric_fields:
    crime_df[field] = pd.to_numeric(crime_df[field])

crime_df.dtypes
```

The `OCCURRED_ON_DATE` field is also an object whereas we require Timestamp object to work with temporal events. We can use python libraries to create datetime objects for this column. Lets start by taking a look at the format.

```python
# Look at the temporal information in the dataframe
crime_df.loc[:10, ('HOUR', 'MONTH', 'YEAR', 'OCCURRED_ON_DATE')]
```

#### Create timestamp field
The dates from the API are in format `2017-03-08T21:34:00` (YYYY-MM-DDTHH:MM:SS). 

For spatial-temporal analysis with arcpy, we need to process this format of date into a `Timestamp` object as a new field in the DataFrame.

```python
import datetime

def process_date(input_date):
    '''Function to process date in YYYY-MM-DDTHH:MM:SS format'''
    try:
        mydate = datetime.datetime.strptime(input_date, '%Y-%m-%dT%H:%M:%S')
    except:
        mydate = datetime.datetime.strptime('1700', '%Y')
    return mydate
```

```python
# Calculate the new time column
crime_df['DateTime'] = crime_df.apply((lambda x: process_date(x['OCCURRED_ON_DATE'])), axis=1)
crime_df.loc[:10, ('OCCURRED_ON_DATE', 'DateTime')]
```

```python
# Confirm that the new field is the correct Timestamp object we expect
print(type(crime_df.iloc[0]['DateTime']))
crime_df.dtypes
```

#### Examine the coordinates

If any of the coordinates are missing we will not be able to create a point for the incident. We will use the dataframe to interrogate the coordinates column and look for missing or odd values.

```python
# Check if there are any invalid Lat and Long that can't be plot on to map
crime_df.loc[:, ('OBJECTID', 'Lat', 'Long')].head(15)
```

There are NaNs, and for Boston a (long,lat) of (-1.0, -1.0) is an invalid coordinate. We will need to filter this out of the data set we use to create points.

We can now use a query on the data frame to remove null and -1 values. A property of NaN values is that they do not return true when compared to themselves, so we can use that in our query. We also filter out -1 values. 

```python
# Remove locations with no coordinates
crime_clean_df = crime_df.query('(Long == Long) and (Long != -1)').copy()
# Print out the first 15 records to check if invalid entries have been removed correctly
crime_clean_df.loc[:, ('OBJECTID', 'Lat', 'Long')].head(15)
```

```python
cleaned_number_of_records = len(crime_clean_df)
```

```python
percent_records_dropped = (original_number_of_records - cleaned_number_of_records)/(cleaned_number_of_records+original_number_of_records) * 100
print("From the API we recieved {} records. After cleaning {} "\
      "records remained. This works out to {:.2f}% of data dropped during cleaning.".format(original_number_of_records, cleaned_number_of_records,
                                                                                       percent_records_dropped))
```

## Time Series Clustering

The [Time Series Clustering](http://pro.arcgis.com/en/pro-app/tool-reference/space-time-pattern-mining/time-series-clustering.htm) tool partitions a collection of time series, stored in a space-time cube, based on the similarity of time series characteristics. Time series can be clustered so they have similar values in time or similar behaviors or profiles across time (increase or decrease at the same points in time).

By clustering on the incident time series areas with similar temporal patterns can be identified. As an example if some precincts had the most activity in the morning hours and others had the most activity in the evenings it would create two distinct clusters. By understanding the temporal patterns police efforts can be timed to maximize their efficiency. There are two modes for time series clustering: VALUE and PROFILE. By using VALUE, locations that have time series with similar values of the Analysis Variable at the same points in time will be clustered together. By using PROFILE, time series that have similar behaviors, fluctuations, and complexity will be clustered together. Clustering is based on the common form or shape of each location's time series. We will examine using the value mode in this analysis to find clusters with similar number of incidents.

```python
# Create output path for Time Series Clustering results
time_series_clustering = os.path.join(analysis_gdb, 'assault_time_series_clustering')

# Run time series clustering based on VALUE
time_series_clustering_results = arcpy.stpm.TimeSeriesClustering(crime_nc, "COUNT", time_series_clustering, "VALUE")

print(time_series_clustering_results.getMessages())
```

### Examine the Time Series Trends

We have examined the locations of the time series clusters. Now we will take a look at what the actual temporal patterns are throughout the year for each of the clusters. We will groupby using the cluster ID and then calculate the minimum, maximum and mean value for the time series and use the values in a plot.

We will merge the time series clusteirng results using CLUSTER_ID with the aggregated point count data so we can look at the results
from each of the clusters

```python
# Join the cluster ID to each location in the visualize 3D results.
join_df = crime_viz3d_sedf.merge(time_series_clustering_sedf, on='LOCATION')

# Group by the start date and the cluster ID so we can get the number of crimes for each month for each cluster
gdf = join_df.groupby(['START_DATE', 'CLUSTER_ID'])

# Get the number of cluster ids
num_clusters = gdf['VALUE'].max().reset_index()['CLUSTER_ID'].unique()
print("There are {} time series clusters for each month.".format(max(num_clusters)))

# Calculate summary statistics for each cluster
max_values = gdf['VALUE'].max().reset_index()
max_values.head(6)
```

```python
min_values = gdf['VALUE'].min().reset_index()
min_values.head(6)
```

```python
mean_values = gdf['VALUE'].mean().reset_index()
mean_values.head(6)
```

In this plot we will be using both a line plot as well as a [patch](https://bokeh.pydata.org/en/latest/docs/reference/models/glyphs/patches.html).  
Basically patch will be the range for each month, and the line will be the mean for each month.

```python
plot_data_list = []

for cur_cluster in num_clusters:
    print("Working on cluster {}".format(cur_cluster))
    cur_min_values = min_values.query("CLUSTER_ID == {}".format(cur_cluster))['VALUE'].tolist()
    cur_max_values = max_values.query("CLUSTER_ID == {}".format(cur_cluster))['VALUE'].tolist()
    cur_mean_values = mean_values.query("CLUSTER_ID == {}".format(cur_cluster))['VALUE'].tolist()
    
    time_series_length = len(cur_max_values)
    patch_x = list(range(0, time_series_length, 1)) + list(range(time_series_length-1, -1, -1))
    patch_y = cur_min_values + cur_max_values[::-1]
    # Repeat the first point on the patch
    patch_x.append(patch_x[0])
    patch_y.append(patch_y[0])
    
    mean_x = list(range(0, time_series_length, 1))
    mean_y = [round(num,2) for num in cur_mean_values] # round to 2 decimal points
    
    plot_data_list.append({'patch_x': patch_x,
                      'patch_y': patch_y,
                      'mean_x': mean_x,
                      'mean_y': mean_y})

# Check the result
for plot_data in plot_data_list:
    print("\n")
    for key in plot_data.keys():
        print(key, ":", plot_data[key])
```

```python
# Establish plotting colors
plotting_colors = ['#78AAFF', '#FF6455', '#7DDC55', '#FFB400', '#C864E1', '#BEA064', '#FABEC8', 
          '#AFAFAF', '#005AE6', '#E60000', '#37A000', '#960096', '#B4FF00', '#822800', 
          '#3C6E82', '#FF00C3', '#00E6AA', '#FFE600', '#002378', '#D78787', '#282828', 
          '#73E1E1', '#006400', '#E1C3FF', '#966432', '#FFC88C', '#D2FFBE', '#CDE1FF', 
          '#FFFF87', '#F0F0F0', ]

p = figure(plot_height=350, plot_width=800)

for idx, cur_plot in enumerate(plot_data_list):
    p.patch(cur_plot['patch_x'], cur_plot['patch_y'], alpha=0.1, color=plotting_colors[idx], line_dash='dashed', line_width=4)
    p.line(cur_plot['mean_x'], cur_plot['mean_y'], color=plotting_colors[idx], line_width=2, legend="Cluster "+str(idx+1))

p.xaxis.axis_label = "Time Step - Monthly"
p.yaxis.axis_label = "Incident Count"

show(p)
```

```python
# Set up output paths for the analysis
cur_crime_points = os.path.join(analysis_gdb, 'cur_assault_points')
precincts_feature = os.path.join(analysis_gdb, 'precincts_feature')
cur_hotspot_result = os.path.join(analysis_gdb, 'cur_hotspot_result')
cur_crime_viz3d = os.path.join(analysis_gdb, 'cur_assault_viz3d')
cur_crime_nc = os.path.join(outputSpace, "cur_crime_cube.nc")

# Write the current crime points to a feature class for analysis
cur_crime_sedf.spatial.to_featureclass(cur_crime_points)
```

```python
# Aggregate into space time cube
arcpy.stpm.CreateSpaceTimeCube(cur_crime_points, cur_crime_nc,
                               "DateTime", None, "1 Weeks", "END_TIME", None, None, None,
                               "DEFINED_LOCATIONS", precincts_feature, "PrecinctWardID")
                               
viz3d_results = arcpy.stpm.VisualizeSpaceTimeCube3D(cur_crime_nc, "COUNT", "VALUE", cur_crime_viz3d)

# Load results to a data frame
cur_crime_viz3d_sedf = pd.DataFrame.spatial.from_featureclass(cur_crime_viz3d)                              
```

```python
# Get the last four weeks of data for this analysis
last_four_weeks = list(cur_crime_viz3d_sedf['START_DATE'].unique())[-4:]
```

```python
# Collect the top 5 precincts for crime incidents for the last four weeks

count_list = []
precinct_list = []
date_list = []

for cur_week in last_four_weeks:
    # Loop through the last four weeks and determine if any of the current week top precints
    # were in the previous week
    query_df = cur_crime_viz3d_sedf[cur_crime_viz3d_sedf['START_DATE'] == cur_week]
    cur_crime_by_precinct = query_df.groupby('PRECINCTWARDID')['VALUE'].sum().sort_values(ascending=False)
    count_list = count_list + list(cur_crime_by_precinct[:5].values)
    precinct_list = precinct_list + list(cur_crime_by_precinct.index[:5])
    date_list = date_list + [cur_week] * 5
    
cur_crime_dict = {"PRECINCT": precinct_list,
             "COUNT": count_list,
                 "DATE":date_list}
```

```python
# Create a data frame with the crime count by precinct for the last four weeks
crime_weeks_df = pd.DataFrame(cur_crime_dict)
# Pull the most recent date from the data
most_recent_date = max(crime_weeks_df['DATE'].unique())
# Format the date for display
time_string = pd.to_datetime(str(most_recent_date)) .strftime('%Y.%m.%d')
# Create a subset data frame with the most recent week
recent_week_df = crime_weeks_df[crime_weeks_df['DATE']==most_recent_date]
# Create a subset data frame with the previous 3 weeks
previous_week_df = crime_weeks_df[crime_weeks_df['DATE']!=most_recent_date]
```

```python
report_string = ""

for cur_precinct in recent_week_df['PRECINCT']:
    """Loop thorugh each precinct in the top 5 for the most recent week and compare to
    the previous 3 weeks to determine if it is new in the top 5 or a repeat"""
    
    # Group the previous 3 weeks by precint
    previous_week_precincts = previous_week_df.groupby('PRECINCT').count()
    
    # Check if the current precinct was in the top 5 previously
    if cur_precinct in previous_week_precincts.index:
        
        cur_index = list(previous_week_precincts.index).index(cur_precinct)
        # Count how many times this precinct has been in the top 5 in the last 4 weeks
        cur_number_of_times = int(previous_week_precincts.iloc[cur_index]['COUNT'])
        cur_string = "Precinct {} has been in the top 5 for crimes for the past {} week(s)\n".format(cur_precinct, cur_number_of_times)
    
    else:
        cur_string = "Precinct {} is new in the top 5. Action required\n".format(cur_precinct)
        
    report_string = report_string + cur_string

# Determine the crime incident trend for the past 4 weeks
total_crimes_by_week = crime_weeks_df.groupby('DATE')['COUNT'].sum().values        
crime_slope = linregress([0, 1, 2, 3], total_crimes_by_week).slope

if crime_slope < 0:
    cur_string = "Crime incidents are on the decrease. Send our regards to the Police Chief\n"
else:
    cur_string = "Crimes incidents are on the rise. Set a meeting with the Police Chief\n"
    
report_string = report_string + cur_string
print(report_string)
```
